```python
df = pd.read_csv("../data/processed/cleaned_covid_data.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
top_countries = (
    df.groupby("Country/Region")["Confirmed"]
      .max()
      .sort_values(ascending=False)
      .head(10)
)

top_countries
```

```python
top_countries.plot(kind="bar")

plt.title("Top 10 Countries by Confirmed Cases")
plt.xlabel("Country")
plt.ylabel("Confirmed Cases")

plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

daily_cases = df.groupby('Date')['Confirmed'].sum()
```

```python
plt.figure(figsize=(12,6))

daily_cases.plot()

plt.title("Global COVID-19 Confirmed Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

daily_cases.plot()

plt.title("Global COVID-19 Confirmed Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")

plt.tight_layout()

plt.savefig("../images/covid_trend_over_time.png")

plt.show()
```

# Key Insights

1. The United States recorded the highest number of confirmed COVID-19 cases.

2. A small number of countries accounted for a large proportion of global deaths.

3. Recovery numbers increased significantly over time, indicating improvements in treatment and management.

4. The global trend chart shows rapid growth in confirmed cases during major outbreak periods.

5. The pandemic impact was uneven across countries, with some regions experiencing significantly higher case counts.

```python
df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']

top_active = (
    df.groupby("Country/Region")["Active"]
      .max()
      .sort_values(ascending=False)
      .head(10)
)

top_active
```
